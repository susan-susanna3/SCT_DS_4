```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
file = list(uploaded.keys())[0]
df = pd.read_csv(file)
```

```python
print("Dataset Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nMissing Values:\n", df.isnull().sum())
```

```python
df.fillna("Unknown", inplace=True)
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Accident_severity", data=df, palette="Set2")
plt.title("Accident Severity Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(y="Weather_conditions", data=df,
              order=df['Weather_conditions'].value_counts().index,
              palette="viridis")
plt.title("Accidents by Weather Conditions")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(y="Road_surface_conditions", data=df,
              order=df['Road_surface_conditions'].value_counts().index,
              palette="coolwarm")
plt.title("Accidents by Road Surface Conditions")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(y="Light_conditions", data=df,
              order=df['Light_conditions'].value_counts().index,
              palette="magma")
plt.title("Accidents by Lighting Conditions")
plt.show()
```

```python
if "Time" in df.columns:
    df["Hour"] = pd.to_datetime(df["Time"], errors="coerce").dt.hour
    plt.figure(figsize=(10,5))
    sns.histplot(df["Hour"].dropna(), bins=24, kde=False, color="skyblue")
    plt.title("Accidents by Hour of the Day")
    plt.xlabel("Hour of Day")
    plt.ylabel("Accident Count")
    plt.show()
```

```python
## Severity by Weather
plt.figure(figsize=(8,5))
sns.countplot(x="Accident_severity", hue="Weather_conditions", data=df, palette="Set3")
plt.title("Accident Severity by Weather")
plt.xticks(rotation=45)
plt.show()
```

```python
# Severity by Road Surface
plt.figure(figsize=(8,5))
sns.countplot(x="Accident_severity", hue="Road_surface_conditions", data=df, palette="Set1")
plt.title("Accident Severity by Road Surface Conditions")
plt.xticks(rotation=45)
plt.show()


```

```python
# Severity by Lighting
plt.figure(figsize=(8,5))
sns.countplot(x="Accident_severity", hue="Light_conditions", data=df, palette="Dark2")
plt.title("Accident Severity by Lighting Conditions")
plt.xticks(rotation=45)
plt.show()
```
